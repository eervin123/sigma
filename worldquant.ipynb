````python
# Import our data
btc_90M_db_vbt = vbt.BinanceData.load('data/btc_90M_db_vbt.pkl')

insample_data = btc_90M_db_vbt['2021-01-01':'2023-01-01']
outofsample_data = btc_90M_db_vbt['2023-01-01':'2023-06-03']
print(insample_data.shape)
print(outofsample_data.shape)

# If you want to work with 1 minute data
data_path = 'data/BTCUSDT_1m.pkl'
min_data = vbt.BinanceData.load(data_path)
print(min_data.shape)
````

Add all of the worldquant alphas in a one liner

````python
alphadf = insample_data.run(["wqa101_%d" % i for i in range(1, 102)], missing_index="drop") # 101 strategies
alphadf
````

````python
# Create a mask of all columns where every value is nan return the columns where every value is nan
mask = alphadf.isna().all(axis=0)
alphadf = alphadf.loc[:, ~mask] # Drop the columns where every value is nan
alphadf
````

````python
# Create a combined dataframe of insample_data and alphadf
insample_df = insample_data.get()
insample_df = pd.concat([insample_df, alphadf], axis=1)
insample_df
````
